```python
import numpy as np
import matplotlib.pylab as plt
from scipy.optimize import minimize
```

```python
np.random.seed(302)
x = np.random.rand(10) * 10
y = np.random.randn(10) + x
```

```python
plt.plot(x, y, '.')
```

```python
def line(a, b, x):
    return a*x + b

def L1(pars, x, y_obs):
    y_pred = line(pars[0], pars[1], x)
    return np.sum(np.abs(y_pred - y_obs))

def L2(pars, x, y_obs):
    y_pred = line(pars[0], pars[1], x)
    return np.sum((y_pred - y_obs)**2)


a0, b0 = 10, -3
pars = minimize(L1, [a0, b0], args=(x, y)).x
print(pars)
plt.plot(x, y, '.')
plt.plot(x, line(pars[0], pars[1], x), '-', label="L1")

pars = minimize(L2, [a0, b0], args=(x, y)).x
print(pars)
plt.plot(x, line(pars[0], pars[1], x), '-', label="L2")
plt.legend()
plt.show()

```
